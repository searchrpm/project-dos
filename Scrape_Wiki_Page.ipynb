```python
# URL of page to be scraped
url = 'https://en.wikipedia.org/wiki/List_of_current_UFC_fighters'
url = 'https://en.wikipedia.org/wiki/List_of_current_UFC_fighters#Debuted_fighters'

# Retrieve page with the requests module
response = requests.get(url)
# Create BeautifulSoup object; parse with 'lxml'
soup = BeautifulSoup(response.text, 'lxml')
```

```python
# Examine the results, then determine element that contains sought info
# results are returned as an iterable list

import csv


table = soup.find_all('table')
t_data = soup.select('table.wikitable')

out = []

for table in t_data:
    #thead = table.select_one('thead th a')
    #print(thead)
    #if not thead:
    #    continue
    
    #if thead.text() != 'ISO':
     #   continue
        
    rows = table.select('tbody tr')
    for row in rows:
        tds = row.select('td')
        if len(tds) == 9:
            country, name, age, ht, nickname, _, _, endeavor, mma = tds
            try:
                country = country.select_one('a')['title'].strip()
            except (TypeError, KeyError):
                continue
            
            name = name.text.strip()
            age = age.text.strip()
            if not age.isdecimal():
                continue
            
            ht = ht.text.strip()
            nickname = nickname.text.strip()
        
            endeavor = endeavor.text.strip()
        
            mma = mma.text.strip()
            
            out.append([country, name, age, ht, nickname, endeavor, mma])
        
        
with open('output.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['Country', 'Name', 'Age', 'Ht', 'Nickname', 'Endeavor', 'MMA'])
    writer.writerows(out)
        
        
with open('output.csv', 'r') as f:
    for line in f: 
        print(line)

        
#for td in t_data:
 #   print(td, end='\n\n\n\n')

#for _a in t_data:
#    try:
#        print(_a['title'])
#    except KeyError:
#        pass
        
#print(t_data)

for td in t_data:
    td.select_one('td > a.thumborder')

fighter_flag = soup.find('td')

#table_rows = table.find('td')

#for fighter in t_data:
    #try:
        #Country = result.find('a', class_='data-sort-value').text


#for td in table:
    #td = table.find('td')
    
# Loop through returned results
#for result in results:
    # Error handling
    #try:
        # Identify and return title of listing
        #Country = result.find('a', class_='result-title').text
        # Identify and return price of listing
        #Name = result.find('span', class_ ='data-sort-value').text
        # Identify and return link to listing
        #Age = result.find('span', class_ ='data-sort-value').text
        # Identify and return link to listing
        #Ht = result.find('td').text
        # Identify and return link to listing
        #Nickname = result.find('i').text
        # Identify and return link to listing
        #Record = result.find('span', class_ ='data-sort-value').text
```
